# Visualization

### Q1. which states are rarely participating for honey production since last 27 years?

```python
df_statecount = df['state'].value_counts()
df_statecount
```

```python
fig , ax = plt.subplots(figsize = (10, 10))
ax = plt.pie(x = df_statecount.values, labels =df_statecount.index , autopct='%1.2f', shadow=True , colors = sns.color_palette('Set2'))
plt.show()
```

## Q2. Which are the top 5 Honey producing state?

```python
df
```

```python
highestprod = df.groupby('state').sum()['production'].reset_index()
highestprod = highestprod.sort_values(by = ['production'], ascending=False)
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.barplot(data = highestprod , x='state' , y= 'production')
plt.xticks(rotation = 90)
plt.show()
```

## 3.What is the Change in mean Average price of Honey 

```python
Changeinavgprice = df.groupby('year')['average_price'].mean().reset_index()
Changeinavgprice
```

```python
fig, ax = plt.subplots(figsize= (15, 8))
ax = plt.plot(Changeinavgprice['year'] ,Changeinavgprice['average_price'], marker = 'o', linewidth = 2 , color = 'purple', label = 'Average price of Honey')
plt.legend()
plt.title('Average Price per year')
plt.show()
```

## Which was the year when production of Honey in whole US was the highest?

```python
df
```

```python
highprod = df.groupby('year')['production'].sum().reset_index()
highprod = highprod.sort_values(by = ['production'], ascending=False)
highprod
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.barplot(data = highprod, x = 'year' , y = 'production')
plt.xticks(rotation = 45)
plt.show()
```

##  Q5.from the above inference we get the production was highest in the year 2000, now let infer which state was having highest contribution in that year?

```python
coly2000 = df['year'] == 2000
```

```python
dfyear2000 = df[coly2000]
dfyear2000
```

```python

hprod2000 = dfyear2000.sort_values(by = ['production'], ascending = False)
hprod2000
```

```python
hprod2000.head(1)
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.barplot(data = hprod2000, x = 'state' , y = 'production')
plt.xticks(rotation = 90)
plt.title('Highest Production')
plt.show()
```

##  Q6.Which states are having the highest no. of colonies in year 2000?

```python
highcol2000 = dfyear2000.sort_values(by = ['colonies_number'], ascending = False)
highcol2000
```

```python
highcol2000['state'].iloc[0]
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
ax = sns.barplot(data = highcol2000 , x = 'state' , y = 'colonies_number')
plt.xticks(rotation = 90)
plt.title('Highest number of Colonies')
plt.show()
```
